```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv("raw/data.csv")    # change filename if needed

print(df.shape)
df
```

```python
df = df.drop(columns=["Unnamed"])
```

```python
df["binary_y"] = np.where(df["y"] == 1, 1, 0)
```

```python
X = df.iloc[:, :-2]      # X1..X178   (exclude y and binary_y)
y = df["binary_y"]       # new binary target column
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization

model = Sequential([
    Dense(256, activation='relu', input_shape=(178,)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(128, activation='relu'),
    Dropout(0.3),

    Dense(64, activation='relu'),
    Dropout(0.2),

    Dense(1, activation='sigmoid')   # Binary classification
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    epochs=35,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = (model.predict(X_test) > 0.5).astype(int)

print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
model.save("seizure_binary_model.h5")
import joblib
joblib.dump(scaler, "scaler.pkl")

print("Model + scaler saved ✅")
```

```python
def predict_eeg(eeg_values):
    eeg_values = np.array(eeg_values).reshape(1, -1)
    eeg_values = scaler.transform(eeg_values)
    pred = (model.predict(eeg_values) > 0.5).astype(int)

    return "SEIZURE DETECTED ⚠️" if pred == 1 else "Normal EEG ✅"
```

```python
sample = df.iloc[1, :-2].values
result = predict_eeg(sample)
print("Row result:", result)
```
